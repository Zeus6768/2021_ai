## (7) 임의의 함수를 정하고 이를 미분한 함수를 같이 만드세요. 이 둘을 한개의 그래프에 같이 표시하고 미분값이 의미하는 바를 적어보세요.

```python
x = np.arange(-6, 11)
f = x ** 2 - 4 * x + 4
g = 2 * x - 4
plt.plot(x, f)
plt.plot(x, g, linestyle=":")
plt.show()
```

임의의 함수는 f(x), 미분한 함수는 g(x)이다. 

미분값은 x의 변화에 따른 각 함수의 결과값의 변화량을 의미한다. 

예를들어 f(2)는 극소점이고 극소점에서 변화량은 0이므로 g(2)의 값 또한 0이 된다. 

다른 예로 f(x) 위의 한 점 (3, 1)에서는 g(x)의 결과값이 2가 된다.
